# Model analysis

```python
# training dataset
df = train.loc[:, train.columns != 'SalePrice']
y = train.loc[:, train.columns == 'SalePrice']
```

#### Train/test split

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2)
```

```python
data=[X_train,y_train]
corrmat = train.corr() 
cols = corrmat.nlargest(6, 'SalePrice')['SalePrice'].index
cols=cols[1:]
```

```python
X_train = X_train[cols]
X_test = X_test[cols]
```

```python
from sklearn import datasets, linear_model
```

```python
test[cols].isnull().sum()
```

```python
test['GarageCars'].fillna(0, inplace=True)
```

```python
# fit a model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
real_prediction = lm.predict(test[cols])
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
```

```python
from sklearn.metrics import mean_squared_log_error
mean_squared_log_error(y_test, real_prediction)  
```

#### Plot of residuals

```python
from yellowbrick.regressor import ResidualsPlot
```

```python
visualizer = ResidualsPlot(model)
visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.poof()                 # Draw/show/poof the data
```

#### K-fold cross validation

```python
from sklearn.model_selection import train_test_split, cross_val_score, validation_curve
```

```python
scores = cross_val_score(model, X_train, y_train, cv=5)
print(scores)
```
